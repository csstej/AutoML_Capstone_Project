```python
df=hyperband2
# df
```

```python

mean_val_loss = df['val_Loss'].mean()

df = df[df['val_Loss'] < mean_val_loss]

# filtered_df = filtered_df[column_name].sort_values(ascending=True)
df.index
```

```python
# df=df.sort_index()
df=df.sort_values(by=['val_Loss'])
df.index
```

```python
import matplotlib.pyplot as plt

# mean_val_loss = df['val_Loss'].mean()


# Create a bar plot of val_Loss
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['val_Loss'])

# for x, y in zip(df.index, df['val_Loss']):
#     plt.plot([x, x], [mean_val_loss, y], color='gray')



plt.title('Index vs. Validation Loss')
plt.xlabel('Index')
plt.ylabel('Validation Loss')

# Mark the mean value on the plot
plt.axhline(y=mean_val_loss, color='r', linestyle='--', label=f'Mean Val_Loss = {mean_val_loss:.2f}')
plt.legend()

plt.grid(True)
plt.show()
```
