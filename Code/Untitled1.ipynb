```python
import matplotlib.pyplot as plt
import numpy as np

# Existing DataFrame 'df' with columns '1pc', '2pc', '3pc', '4pc', and '5pc'

# Boxplot
df.boxplot(column=['1pc', '2pc', '3pc', '4pc', '5pc'], showfliers=True)

# Add individual points
colors = ['blue', 'green', 'red', 'purple', 'orange']
labels = ['1pc', '2pc', '3pc', '4pc', '5pc']
for i, col in enumerate(labels):
    y = df[col]
    x = np.random.normal(i+1, 0.04, size=len(y))  # Add some jitter to x coordinates for better visibility
    plt.scatter(x, y, alpha=0.7, color=colors[i], label=col)

plt.xticks(range(1, len(labels)+1), labels)  # Ensure x-ticks match the boxplots
plt.title('Comparison of 1pc, 2pc, 3pc, 4pc, and 5pc')
plt.grid(True)
plt.show()
```
