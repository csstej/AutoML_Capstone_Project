```python
best_architecture_row = df.sort_values(by='val_Loss').iloc[0]

# Extract architecture configurations
layer1_units = best_architecture_row['layer_block_1/Layer_1']
activation1 = best_architecture_row['layer_block_1/activation_1']
layer2_units = best_architecture_row['layer_block_1/Layer_2']
activation2 = best_architecture_row['layer_block_1/activation_2']
dropout_rate = best_architecture_row['classification_head_1/dropout']
optimizer = best_architecture_row['optimizer']
learning_rate = best_architecture_row['learning_rate']

# Define a function to create a model based on architecture configurations
def create_model():
    model = keras.Sequential([
        keras.layers.Dense(layer1_units, activation=activation1),
        keras.layers.Dense(layer2_units, activation=activation2),
        keras.layers.Dense(1)  # Output layer for regression
    ])

    model.compile(optimizer=optimizer, loss='mean_squared_error', metrics=['mae'])  # Adjust loss and metrics as needed

    return model


# Create the best model based on the extracted architecture configurations
best_model = create_model()

# Print the best architecture configurations
print("Best Architecture Configurations:")
print(f"Layer 1 Units: {layer1_units}")
print(f"Activation 1: {activation1}")
print(f"Layer 2 Units: {layer2_units}")
print(f"Activation 2: {activation2}")
print(f"Dropout Rate: {dropout_rate}")
print(f"Optimizer: {optimizer}")
print(f"Learning Rate: {learning_rate}")
```
